```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

from datetime import datetime
```

```python
from yahoofinancials import YahooFinancials


yahoo_financials = YahooFinancials('MSFT')
data = yahoo_financials.get_historical_price_data(start_date='2000-01-01', 
                                                   end_date='2022-12-31', 
                                                   time_interval='weekly')
 
msft_df = pd.DataFrame(data['MSFT']['prices'])
 
msft_df = msft_df.drop('date', axis=1).set_index('formatted_date')
msft_df.head(100)
# msft_df.set_index('date')
```
